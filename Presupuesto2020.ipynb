### Datos
- Presupuesto de gastos y su ejecución detallada - agrupación mensual 2020
- Importados de https://datos.gob.ar/dataset/sspre-presupuesto-administracion-publica-nacional-2020/archivo/sspre_275

```python
data = pd.read_csv('credito-mensual-2020.csv')
print(data.columns)
```

```python
data.head()
```

#### Filtro la Base respecto al Ministerio de Salud

```python
data = data.loc[data.jurisdiccion_desc=='Ministerio de Salud']
```

```python
provincias = list(data.ubicacion_geografica_desc.unique())
indices    = [6,7]
provincias = [i for j, i in enumerate(provincias) if j not in indices]
print(provincias)
print('\n')
print(len(provincias))
```

#### Dataset Cleaning
- Hay valores que tienen "," que no pueden ser operados matemáticamente
- Elementos string necesitan ser reemplazados por Elementos float

```python
credito_pres = []
credito_vig = []
credito_dev = []

for i in range(0,len(data)):
    
    if isinstance(data.credito_presupuestado.iloc[i], str):
        flotante = float(data.credito_presupuestado.iloc[i].replace(',','.'))
        credito_pres.append(flotante)
    else: 
        entero = data.credito_presupuestado.iloc[i]
        credito_pres.append(entero)
        
    if isinstance(data.credito_vigente.iloc[i], str):
        flotante = float(data.credito_vigente.iloc[i].replace(',','.'))
        credito_vig.append(flotante)
    else: 
        entero = data.credito_vigente.iloc[i]
        credito_vig.append(entero)   
        
    if isinstance(data.credito_devengado.iloc[i], str):
        flotante = float(data.credito_devengado.iloc[i].replace(',','.'))
        credito_dev.append(flotante)
    else: 
        entero = data.credito_devengado.iloc[i]
        credito_dev.append(entero)     
      
```

#### Datos de interés

```python
credito_pres = pd.DataFrame(credito_pres)
credito_vig  = pd.DataFrame(credito_vig)
credito_dev  = pd.DataFrame(credito_dev)
ubicacion    = pd.DataFrame(data.ubicacion_geografica_desc.values)
```

#### Tabla Final

```python
creditos = pd.concat([ubicacion,credito_pres,credito_vig,credito_dev],axis=1)
creditos.columns = ['Provincia','Credito_Presupuestado','Credito_Vigente','Credito_Devengado']
creditos
```

#### Datos Agrupados por Provincias

```python
creditos.groupby('Provincia').sum().sort_values('Credito_Presupuestado',ascending=False)
```
